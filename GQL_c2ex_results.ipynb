# c2ex_results

```python
import requests
import json
import pandas as pd
GRAPHQL_URL = 'http://54.166.100.68:8080/v1/graphql'
```

```python
body = """
{
  c2ex_results {
    systemid
    dollarloss
    dollarwin
    numtrades
}}"""
r = requests.post(GRAPHQL_URL, json={'query': body})
json_data = json.loads(r.text)
df_data = json_data["data"]["c2ex_results"]
df = pd.DataFrame(df_data)
display(df)
```

```python
body = """
{
  c2ex_results(where: {systemid: {_eq: "126054331"}}) {
    systemid
    dollarloss
    dollarwin
    numtrades}}"""
r = requests.post(GRAPHQL_URL, json={'query': body})
#print(r.status_code)
#print(r.text)
json_data = json.loads(r.text)
df_data = json_data["data"]["c2ex_results"]
df = pd.DataFrame(df_data)
display(df)
```
