# c2ex_equity_daily

```python
import requests
import json
import pandas as pd
GRAPHQL_URL = 'http://54.166.100.68:8080/v1/graphql'
```

```python
body = """
query MyQuery {
  c2ex_equity_daily(limit: 25) {
    datetime
    systemid
    totalpl   
  }
}
"""
r = requests.post(GRAPHQL_URL, json={'query': body})
json_data = json.loads(r.text)
df = pd.DataFrame(json_data["data"]["c2ex_equity_daily"])
display(df)
```

```python
body = """
query MyQuery {
c2ex_equity_daily(limit: 25, where: {systemid: {_in: [125624499, 136287221,133472881]}}) {
    systemid
    datetime
    totalpl   
  }
}
"""
r = requests.post(GRAPHQL_URL, json={'query': body})
json_data = json.loads(r.text)
df = pd.DataFrame(json_data["data"]["c2ex_equity_daily"])
display(df)
```
