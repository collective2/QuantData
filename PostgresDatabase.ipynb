-- public.expectancyview source

CREATE MATERIALIZED VIEW public.expectancyview
TABLESPACE pg_default
AS SELECT sub.systemid,
    sub.datestart AS date,
    (sub.avgwin / sub.avgloss)::double precision - sub.numloss::double precision / sub.numtrades::double precision AS statval
   FROM ( SELECT r.systemid,
            r.datestart,
            ( SELECT count(*) AS count
                   FROM c2ex_trades
                  WHERE c2ex_trades.systemid = r.systemid AND date(c2ex_trades.exittime) <= r.datestart) AS numtrades,
            ( SELECT count(*) AS count
                   FROM c2ex_trades
                  WHERE c2ex_trades.systemid = r.systemid AND date(c2ex_trades.exittime) <= r.datestart AND c2ex_trades.result <= 0::numeric) AS numloss,
            ( SELECT avg(c2ex_trades.result) AS avg
                   FROM c2ex_trades
                  WHERE c2ex_trades.systemid = r.systemid AND date(c2ex_trades.exittime) <= r.datestart AND c2ex_trades.result > 0::numeric) AS avgwin,
            ( SELECT abs(avg(c2ex_trades.result)) AS abs
                   FROM c2ex_trades
                  WHERE c2ex_trades.systemid = r.systemid AND date(c2ex_trades.exittime) <= r.datestart AND c2ex_trades.result <= 0::numeric) AS avgloss
           FROM returnsdatainintervalscleanedskip090 r
             JOIN c2ex_results res ON res.systemid = r.systemid) sub
  WHERE sub.numtrades > 0 AND sub.avgloss > 0::numeric
WITH DATA;
