# historicalstats01

```python
import requests
import json
import pandas as pd
GRAPHQL_URL = 'http://54.166.100.68:8080/v1/graphql'
```

```python
body = """
query MyQuery {
  historicalstats01(where: {agedays: {_gt: 3000}}) {
    date
    strategyid
    agedays
    avgtradeloss
    avgtradewin
    c2star
    calmar
    deltaequity90days
    expectancy
  }
}
"""
r = requests.post(GRAPHQL_URL, json={'query': body})
json_data = json.loads(r.text)
df_data = json_data["data"]["historicalstats01"]
df = pd.DataFrame(df_data)
df = df.sort_values(by=["date", "strategyid"])
display(df)
```
