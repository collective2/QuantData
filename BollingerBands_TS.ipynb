# Bollinger bands in Node.js (TypeScript)

```typescript
import Plotly from "tslab-plotly";
import * as tslab from "tslab";
const postgres = require('postgres');
var BB = require('technicalindicators').BollingerBands;
```

```typescript
let db = postgres('postgres://c2public:c2public@quantdata.collective2.com');
console.log('Connected');
```

```typescript
// Get data for the strategy 117734561
const resultArray = await db`select datetime, totalpl from c2ex_equity_daily where systemid=117734561;`
console.log('Data ready');
```

```typescript
// Prepare time series and plotting options
const dates =  resultArray.map( item => item.datetime);
const equity117734561 = resultArray.map( item => item.totalpl);

var input = {
    period : 30, 
    values : equity117734561 ,
    stdDev : 3    
};

const bbData = BB.calculate(input);
```

```typescript
// Format data for Plotly
const upper = bbData.map( item => item.upper);
const middle = equity117734561;
const lower = bbData.map( item => item.lower);

const upperTrace = {x:dates, y:upper, mode:'lines', name: 'UpperBB'};
const middleTrace = {x:dates, y:middle, mode:'lines', name: 'Equity'};
const lowerTrace = {x:dates, y:lower, mode:'lines', name: 'LowerBB'};

const data = [upperTrace,middleTrace,lowerTrace];
```

```typescript
var layout = {autosize: false,width: 800,height: 600};
Plotly.newPlot(tslab, data,layout);
```
