# c2systems

```python
import requests
import json
import pandas as pd
GRAPHQL_URL = 'http://54.166.100.68:8080/v1/graphql'
```

```python
body = """
query MyQuery {
  c2systems {
    systemid
    added
    alive
    cash
    closedpl
    equity
    openpositionspl
    startingcash
    systemname
  }
}
"""
r = requests.post(GRAPHQL_URL, json={'query': body})
json_data = json.loads(r.text)
df = pd.DataFrame(json_data["data"]["c2systems"])
display(df)
```

```python
body = """
{
  c2systems (where: {systemid: {_eq: "13202557"}}) {
    systemid
    added
    alive
    cash
    closedpl
    equity
    openpositionspl
    startingcash
    systemname
  }
}
"""
r = requests.post(GRAPHQL_URL, json={'query': body})
json_data = json.loads(r.text)
# print(json_data)
df = pd.DataFrame(json_data["data"]["c2systems"])
display(df)
```
